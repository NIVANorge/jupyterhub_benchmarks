```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn

plt.style.use("ggplot")
```

# JupyterHub benchmarking

## Plot results

```python
xl_path = r"../results/hub_benchmarks.xlsx"
df = pd.read_excel(xl_path)
df.dropna(how='any', inplace=True)
df.head()
```

```python
sn.catplot(
    data=df,
    x="Hub",
    y="Value",
    col="Name",
    kind="bar",
    col_wrap=3,
    aspect=1.5,
    height=3,
    sharex=False,
    sharey=False,
    legend_out=False,
)

png_path = r"../results/test_results.png"
plt.savefig(png_path, dpi=200)
```
